```python
class Node:
    def __init__(self):
        self.val = None
        self.left = Node()
        self.right = Node()
```

```python
class BinaryTree:
    def __init__(self):
        self.head = None
    def add(self,num,head=self.head):
        node = Node()
        node.left = None
        node.right = None
        node.val = num
        if not head:
            head = node
        elif not head.left:
            self.add(head.left)
        else:
             self.add(head.right)
```

```python
mytree = BinaryTree()
mytree.add(1)
```
